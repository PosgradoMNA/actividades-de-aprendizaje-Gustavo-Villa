```python
import pandas as pd
import numpy as np
```

```python
#descargar tabla y crear dataframe
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(path, header=None)
```

```python
#Ahora mostraremos las primeras 5 filas de nuestro df

print("The first 5 rows of the dataframe") 
df.head(5)
```

```python
#Ahora mostraremos las primeras 10 filas de nuestro df

print("The first 10 rows of the dataframe") 
df.head(10)
```

```python
# Creamos los títulos para nuestro dataframe
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
#Asignamos los títulos y validamos
df.columns = headers
df.head(10)
```

```python
#Empezamos la limpieza sustituyendo "?" por "NaN"
df1=df.replace('?',np.NaN)
```

```python
#Soltamos los valores faltantes de la columna price
df=df1.dropna(subset=["price"], axis=0)
df.head(20)
```

```python
# Ahora imprimimos el nombre de las columnas de nuestro dataframe
print(df.columns)
```

```python
# Guardamos nuestro dataframe como CSV
df.to_csv("automobile.csv", index=False)
```

```python
# Revisamos el tipo de nuestras variables
df.dtypes
```

```python
#Generamos un resumen estadístico COMPLETO de nuestro dataframe
df.describe(include = 'all')
```
