```python
#Estimamos la media para la variable Age
wm = ndf.Age.mean()
wm
```

```python
#Imputamos usando la media en nuestra variable Age
ndf['Age'].fillna(value = wm, 
                    inplace = True)
ndf
```

```python
#Validamos que no haya NAs en nuestra variable Age
ndf['Age'].isna().any()
```

```python
#Imputamos ahora nuestra variable Education con base en la mediana, hace sentido al ser una variable categórica
ndf['Education'].fillna(value = ndf.Education.median(), 
                    inplace = True)
ndf
```

```python
#Validamos que no haya NAs en nuestra variable Education
ndf['Education'].isna().any()
```

```python
#Imputamos ahora nuestra variable Marital_Status con base en la moda, hace sentido al ser una variable categórica
mm  = ndf.Marital_Status.mode()
ndf['Marital_Status'].fillna(value = mm,  
                    inplace = True)
ndf
```

```python
#Empezamos de nuevo
ndf = df.copy()
ndf
```

```python
#Ahora soltaremos los NAs dentro de nuestras variables Age y Education

ndf.dropna(subset=['Age', 'Education'], inplace = True)
ndf
```
