```python
#Importación de paqueterías
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
#descargar tabla y crear dataframe
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(path, header=None)
```

```python
# Creamos los títulos para nuestro dataframe
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
#Asignamos los títulos y validamos
df.columns = headers
df.head(10)
```

```python
#Empezamos la limpieza sustituyendo "?" por "NaN"
df1=df.replace('?',np.NaN)
```

```python
#Soltamos los valores faltantes de la columna price
df=df1.dropna(subset=["price"], axis=0)
df.head(20)
```

```python
#Hacemos un subconjunto de datos faltantes
missing_data = df.isnull()
missing_data.head(5)
```

```python
# Contamos los valores faltantes
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")   
```

```python
# Calculamos la media
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# Reemplazamos con la media
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
#Calculamos la media para otra variable
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)

# Reemplazamos con la media
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
# Obtenemos la información general de nuestro dataframe
df.info()
```

```python
#convertimos los formatos en los correctos
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
#transformamos la columna de acuerdo a nuestra necesidad
df["highway-mpg"] = 235/df["highway-mpg"]

df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)

df.head()
```
